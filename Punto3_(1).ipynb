## Tokenización y lematización

```python
%%time
nlp = spacy.load('es_core_news_lg')
stop_words = set(STOP_WORDS)

def process_texts(df):
    sents = []
    tokens = []
    lemma = []
    pos = []
    ner = []

    for doc in nlp.pipe(df['corrected'], batch_size=100, n_process=4):
        if doc.is_parsed:
            tokens.append([token.text for token in doc])
            sents.append([sent.text.strip() for sent in doc.sents])
            pos.append([token.pos_ for token in doc])
            ner.append([ent.text for ent in doc.ents])
            lemma.append([token.lemma_ for token in doc])
        else:
            tokens.append(None)
            lemma.append(None)
            pos.append(None)
            sents.append(None)
            ner.append(None)

    df['tokens'] = tokens
    df['lemma'] = lemma
    df['pos'] = pos
    df['sents'] = sents
    df['ner'] = ner
    return df

datos_train = process_texts(datos_train)
datos_test = process_texts(datos_test)
```

# Ingenieria de variables
Variables al cuadrado

```python

# Definir variables cuantitativas
cuantitativas = ['token_count', 'unique_token_count', 'nostop_count', 'sent_count', 'ner_count',
                 'comma', 'question', 'exclamation', 'quotation', 'organization', 'caps', 'person',
                 'location', 'money', 'time', 'date', 'percent', 'noun', 'adj', 'pron', 'verb',
                 'cconj', 'adv', 'det', 'propn', 'num', 'part', 'intj']

# Aplicar variables al cuadrado
def add_squared_variables(df):
    base_cuadrado = df[cuantitativas].copy()
    for k in cuantitativas:
        base_cuadrado[k + "_2"] = base_cuadrado[k] ** 2
    return base_cuadrado

datos_train_sq = add_squared_variables(datos_train_fe)
datos_test_sq = add_squared_variables(datos_test_fe)
```

Interacciones Cuantitativas

```python
# Aplicar interacciones cuantitativas
def add_interaction_variables(df):
    base_interacciones = df[cuantitativas].copy()
    lista_inter = list(combinations(cuantitativas, 2))
    for k in lista_inter:
        base_interacciones[k[0] + "__" + k[1]] = base_interacciones[k[0]] * base_interacciones[k[1]]
    return base_interacciones

datos_train_inter = add_interaction_variables(datos_train_fe)
datos_test_inter = add_interaction_variables(datos_test_fe)
```

Razones

```python
# Aplicar razones
def add_ratio_variables(df):
    base_razones = df[cuantitativas].copy()
    raz1 = [(x, y) for x in cuantitativas for y in cuantitativas if x != y]
    for j in raz1:
        base_razones[j[0] + "__coc__" + j[1]] = base_razones[j[0]] / (base_razones[j[1]] + 0.01)
    return base_razones

datos_train_ratio = add_ratio_variables(datos_train_fe)
datos_test_ratio = add_ratio_variables(datos_test_fe)
```

```python
# Recolectar resultados
results = {
    'BERT sin ingeniería de variables': roc_auc_bert_no_fe,
    'BERT con ingeniería de variables': roc_auc_bert_fe,
    'FastText sin ingeniería de variables': roc_auc_ft_no_fe,
    'FastText con ingeniería de variables': roc_auc_ft_fe
}

# Identificar el mejor modelo
best_model = max(results, key=results.get)
best_auc = results[best_model]

# Imprimir resultados
print("Resultados de ROC AUC para cada modelo:")
for model, auc in results.items():
    print(f'{model}: ROC AUC = {auc:.4f}')

print(f'\nEl modelo con el máximo valor del área bajo la curva ROC es: {best_model}')
print(f'Con un valor de ROC AUC de: {best_auc:.4f}')
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt

# Generar predicciones para el mejor modelo
if best_model == 'BERT sin ingeniería de variables':
    y_pred_best = y_pred
elif best_model == 'BERT con ingeniería de variables':
    y_pred_best = y_pred_fe
elif best_model == 'FastText sin ingeniería de variables':
    y_pred_best = model.predict_proba(X_test_ft)[:, 1]
else:
    y_pred_best = model.predict_proba(X_test_ft_fe)[:, 1]

# Binarizar las predicciones para la matriz de confusión
threshold = 0.5
y_pred_binary = (y_pred_best >= threshold).astype(int)

# Matriz de confusión
matriz_confusion = pd.DataFrame(confusion_matrix(y_test, y_pred_binary))
plt.figure(figsize=(8, 6))
graph_confusion = sns.heatmap(matriz_confusion, linewidth=0.01, annot=True, cmap='Blues', fmt='.0f')
graph_confusion.set_title('Matriz de confusión para clasificación')
plt.yticks(rotation=0)
plt.xlabel('Valores predichos'); plt.ylabel('Valores observados')
plt.show()

# Curva ROC
print('El área bajo la curva roc es: %0.5f' % best_auc)
fpr, tpr, _ = roc_curve(y_test, y_pred_best)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```
